```python
fig, ax = plt.subplots()
fig.set_size_inches(12,10)

# Store which labels have already been used
used_labels = set()

def plot(days, fluxes, errors, color, label_name):
    for day, flux, err in zip(days, fluxes, errors):
        if flux < 3 * err:
            fmt = 'x'  # square for non-detection
            label = None  # no label for non-detections

            ax.errorbar(day, (flux+3*err), fmt=fmt, color=color, label=label)
            
        if flux > 3*err :
            fmt = '^'  # triangle for detection
            label = label_name if label_name not in used_labels else None
        
            ax.errorbar(day, flux, yerr=err, fmt=fmt, color=color, label=label)
        
        if label:
            used_labels.add(label)

# Plot each band
plot(days1_26, flux1_26, flux_err1_26, 'orange', '1.26 GHz')
#ax.plot(days1_26, flux1_26, color='orange', linestyle = '--', alpha=.50)
plot(days1_74, flux1_74, flux_err1_74, 'red', '1.74 GHz')
#ax.plot(days1_74, flux1_74, color='red', linestyle = '--', alpha=.50)

plot(days4_55, flux4_55, flux_err4_55, 'green', '4.55 GHz')
#ax.plot(days4_55, flux4_55, color='green', linestyle = '--', alpha=.50)
plot(days7_38, flux7_38, flux_err7_38, 'blue', '7.38 GHz')
#ax.plot(days7_38, flux7_38, color='blue', linestyle = '--', alpha=.50)

plot(days13_5, flux13_5, flux_err13_5, 'purple', '13.5 GHz')
#ax.plot(days13_5, flux13_5, color='purple', linestyle = '--', alpha=.50)
plot(days17_4, flux17_4, flux_err17_4, 'hotpink', '17.4 GHz')
#ax.plot(days17_4, flux17_4, color='hotpink', linestyle = '--', alpha=.50)

plot(days28_2, flux28_2, flux_err28_2, 'black', '28.2 GHz')
#ax.plot(days28_2, flux28_2, color='black', linestyle = '--', alpha=.50)
plot(days36_5, flux36_5, flux_err36_5, 'grey', '36.5 GHz')
#ax.plot(days36_5, flux36_5, color='grey', linestyle = '--', alpha=.50)


# Final plot formatting
ax.set_xscale('log')
ax.set_xticks([10,20,50,100,200], labels = ['10','20','50','100','200'], fontsize = 13)
ax.set_yscale('log')
ax.set_yticks([.1,1,10], labels = ['0.1','1.0','10.0'], fontsize = 13)
ax.set_xlabel("Time since eruption (days)", fontsize = 20)
ax.set_ylabel("Flux Density (mJy)", fontsize = 20)

ax.tick_params(axis='both', which='both', direction='in', top=True, right=True)

ax.set_title("")
plt.legend(fontsize = 13)
plt.savefig("V1535Sco_lc.png")
plt.show()
```
