```python

#defining the figure and fig size
fig, ax = plt.subplots()
fig.set_size_inches(12,10)

# Store which labels have already been used
used_labels = set()

#function to sort the non-detections vs the detections
def plot(days, fluxes, errors, color, label_name):
    for day, flux, err in zip(days, fluxes, errors):
        if flux < 3 * err:
            fmt = 'x'  # square for non-detection
            label = None  # no label for non-detections

            ax.errorbar(day, flux, fmt=fmt, color=color, label=label)
            
        else:
            fmt = '^'  # triangle for detection
            label = label_name if label_name not in used_labels else None
        
            ax.errorbar(day, flux, yerr=err, fmt=fmt, color=color, label=label)
        
        if label:
            used_labels.add(label)

# Plot each band
plot(days1_26, flux1_26, flux_err1_26, 'orange', '1.26 GHz')
plot(days1_74, flux1_74, flux_err1_74, 'red', '1.74 GHz')

plot(days5, flux5, flux_err5, 'green', '5 GHz')
plot(days7, flux7, flux_err7, 'blue', '7 GHz')

plot(days13_5, flux13_5, flux_err13_5, 'purple', '13.5 GHz')
plot(days16_5, flux16_5, flux_err16_5, 'hotpink', '16.5 GHz')

plot(days29_5, flux29_5, flux_err29_5, 'black', '29.5 GHz')
plot(days35, flux35, flux_err35, 'grey', '35 GHz')


# Final plot formatting
ax.set_xscale('log')
ax.set_xticks([10,20,50,100,500,1000], labels = ['10','20','50','100','500', '1000'])
ax.set_yscale('log')
ax.set_yticks([.1,1,10,100], labels = ['0.1','1','10', '100'])
ax.set_xlabel("Time since eruption (days)")
ax.set_ylabel("Flux Density (mJy)")

ax.tick_params(axis='both', which='both', direction='in', top=True, right=True)

ax.set_title("")
plt.legend()
plt.show()
```
