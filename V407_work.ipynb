```python

fig, ax = plt.subplots()
fig.set_size_inches(12,10)

# Store which labels have already been used
used_labels = set()

def plot(days, fluxes, errors, color, label_name):
    for day, flux, err in zip(days, fluxes, errors):
        if flux < 3 * err:
            fmt = 'x'  # square for non-detection
            label = None  # no label for non-detections
            ax.errorbar(day, (flux+err*3), fmt=fmt, color=color, label=label)
            
        else:
            fmt = '^'  # triangle for detection
            label = label_name if label_name not in used_labels else None
        
            ax.errorbar(day, flux, yerr=err, fmt=fmt, color=color, label=label)
        
        if label:
            used_labels.add(label)

# Plot each band
plot(days_1_4, flux_1_4, flux_err_1_4, 'orange', '1.4 GHz')
plot(days_1_8, flux_1_8, flux_err_1_8, 'red', '1.8 GHz')
plot(days_4_8, flux_4_8, flux_err_4_8, 'green', '4.8 GHz')
plot(days_7_4, flux_7_4, flux_err_7_4, 'blue', '7.4 GHz')
plot(days_19, flux_19, flux_err_19, 'hotpink', '19 GHz')
plot(days_25, flux_25, flux_err_25, 'teal', '25 GHz')
plot(days_41, flux_41, flux_err_41, 'black', '41 GHz')
plot(days_45, flux_45, flux_err_45, 'yellow', '45 GHz')
plot(days_12_1, flux_12_1, flux_err_12_1, 'purple', '12.1 GHz')
plot(days_142, flux_142, flux_err_142, 'brown', '142 GHz')

# Final plot formatting
ax.set_xscale('log')
ax.set_xticks([10,20,50,100,500,1000], labels = ['10','20','50','100','500','1000'], fontsize =13)
ax.set_yscale('log')
ax.set_yticks([1,10,100], labels = ['1.0','10.0', '100.0'], fontsize =13)
ax.set_xlabel("Time since eruption (days)", fontsize = 20)
ax.set_ylabel("Flux Density (mJy)", fontsize = 20)

ax.tick_params(axis='both', which='both', direction='in', top=True, right=True)

ax.set_title("")
plt.legend(fontsize = 13)
plt.savefig("V407_lc.png")
plt.show()
```
